# Train-test split

Train_test split is a data preparation technique where we divides the dataset into:
1.Training data: used to train the ML model.
2.Testing data: used to evaluate how well the model performs on unseen data.
Why do we need Train-Test Split:
If we train and test on the same data,the model:
- Looks smart
- Fails in real-world data
Train-test split helps to:
-Detect overfitting
-Measure real performance
-Simulate future/unseen data
Real-life analogy:
--
Sudying from a book(train) !=writing exam
questions you've already seen (test)
Common Split Ratios:
-------------------
Training    Testing    When to use

80%           20%       Standard choice
70%           30%       Small dataset
90%           10%       Large dataset

 Example:
 If N=100 rows and split=80:20
 Train=80 samples
 Test=20 samples



```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data=pd.DataFrame({
    "study_Hours":[1,2,3,4,5,6,7,8],
    "Marks":[35,40,50,55,65,70,80,85]
})
    
```

```python
data
```

```python
x=data[["study_Hours"]]
y=data["Marks"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y,test_size=0.20,random_state=55)
```
