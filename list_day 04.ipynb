# File Handling

```python
f=open('abc.txt','w')
print('file name :',f.name)
print('file mode :',f.mode)
print(f.readable())
print(f.writable())
print(f.closed)
f.close()
print(f.closed)
```

# Reading a files

```python
with open("abc.txt","r") as file:
    content=file.read()
print(content)
```

```python
with open("abc.txt","r") as file:
    content=file.read(10)
print(content)
```

```python
with open("abc.txt","r") as file:
    content=file.readline()
print(content)
```

```python
with open("abc.txt","r") as file:
    content=file.readlines()
print(content)
```
