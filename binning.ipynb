# Binning

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.DataFrame({
    "age":[18,22,25,30,35,40,45,50]
})
print(data)
```

```python
data["age_bin"]=pd.cut(data["age"],bins=4)
data
```

```python
plt.hist(data["age"],bins=4)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal-width Binning using pd.cut()")
plt.show()
```

```python
data2=pd.DataFrame({
    "age":[18,22,25,30,35,40,45,50]
})
print(data2)
```

```python
data2["age_bin"]=pd.cut(data2["age"],bins=3)
data2
```

# Equal-Freqency Binning(Quantile Binning):

```python
data2=pd.DataFrame({
    "age":[18,22,25,30,32,35,40,45,50]
})
print(data2)
```

```python
data2["age_frequency_bin"]=pd.qcut(data2["age"], q=4)
data2
```

# Custom/domain-based binning

```python
custom_bins=[17,25,35,50]
labels=["young","Adult","Senior"]
bins=pd.cut(data["age"],bins=custom_bins, labels=labels)
bins
```

```python
plt.hist(bins,bins=4)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal-Frequency Binning using pd.cut()")
plt.show()
```
