# Data pipeline

What is a data pipeline?
-A data pipeline is a step-by-step process that takes raw data and converts it into model-ready data.

Basic Flow:
Raw Data-> Cleaning->processing->Model

Why Do we Need a Data Pipeline?
Real-world/Raw data is not clean:
*Missing Values
*Dulpicate records
*Different formats
*Noise and Outliers

A data pipeline:
*Organizes data properly
*Automates steps
*Reduces human errors
*Makes ML models reliable

Pipeline in sklearn:
-A Pipeline in sklearn is a sequence of data preprocessing steps and machine learning model combined together to automate training and prediction.
DataCollection->Data Preparation->Ml Model Training->Ml model Evaluation->ML model Deployment

# Simple data pipeline
1.Data Collection:
Collecting data from different sources.
Common Sources:
*CSV/Excel files
*API'S
Example:
Student marks data collected from a CSV file.

2.Data Cleaning
Fixing errors and problems in data
Common Tasks:
* Handling missing values
* Removing duplicates
* Fixing incorrect values
* Removing ouliers

Example:
* Fill missing study hours with average
* Remove duplicates study records.

3.Data Preprocessing:
Converting data into a machine-understandable formate.

Common Techniques:
* Scaling(StandardScaler,MinMaxScaler
* Encoding(Label Encoding,One-Hot Encoding)
* Train-Test-Split
Examples:
* Convert "Male/Female"->0/1
* Scale marks to same range
4.Feature Engineering:
  Creating new and better features from existing data.
Common Methods:
* Binning
* Feature Creation
* Feature Selection
* Feature Transformation
Example:
* Convert age->Young/Adult/Senior
* Create total marks from subject marks
 
  5.Model Training:
  Giving  processed data to a machine learning algorithm so it can learn patterns
Common Algorithms:
* Linear Rgression
* Logistic Regression
* Decision-Tree
* Random Forest

6.Model Evaluation:
 Checking how well the model is performing
* Accuracy 
* Mean Squared error(MSE)
* Mean Absolute eror(MAE)
* R Square score

Example:
    Compare prediction  marks with actual marks
