# Train-Test Split

```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data= pd.DataFrame({
    "study_Hours":[1,2,3,4,5,6,7,8],
    "Marks":[35,40,50,55,65,70,80,85]
})
data
```

```python
X = data[["study_Hours"]]
Y = data["Marks"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.20,random_state=40)
```
